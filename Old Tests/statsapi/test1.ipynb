```python
import statsapi
from datetime import datetime

startOfSeason = '2025-03-18'  

# schedule = statsapi.schedule(start_date=startOfSeason, end_date=startOfSeason)
schedule = statsapi.schedule(date=startOfSeason)
display(schedule)

statsapi.schedule
```

```python
if not schedule:
    print(f"No games found for {startOfSeason}")
else:
    for game in schedule:
        game_id = game['game_id']
        home_team = game['home_name']
        away_team = game['away_name']
        game_date = game['game_date']
        
        print(f"\n📅 {away_team} at {home_team} on {game_date}")

        # Step 2: Get boxscore
        box = statsapi.boxscore_data(game_id)

        # Step 3: Loop through teams
        for team_key, team_name in [('away', away_team), ('home', home_team)]:
            team_stats = box.get(team_key, {}).get('teamStats', {})

            batting_stats = team_stats.get('batting', {})
            fielding_stats = team_stats.get('fielding', {})

            runs = batting_stats.get('runs', 0)
            hits = batting_stats.get('hits', 0)
            errors = fielding_stats.get('errors', 0)

            print(f"{team_key.capitalize()} Team: {team_name}")
            print(f"  Runs: {runs}")
            print(f"  Hits: {hits}")
            print(f"  Errors: {errors}")

        print("-" * 50)
```
